Find the points at which two given functions intersect
Consider the example of finding the intersection of a polynomial and a line:

\(y_1=x_1^2\)

\(y_2=x_2+1\)

```python
from scipy.optimize import fsolve

import numpy as np

def f(xy):
   x, y = xy
   z = np.array([y - x**2, y - x - 1.0])
   return z

fsolve(f, [1.0, 2.0])
```

```python
# https://stackoverflow.com/questions/28766692/how-to-find-the-intersection-of-two-graphs
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 1000)
f = np.arange(0, 1000)
g = np.sin(np.arange(0, 10, 0.01) * 2) * 1000

plt.plot(x, f, '-')
plt.plot(x, g, '-')

idx = np.argwhere(np.diff(np.sign(f - g))).flatten()
plt.plot(x[idx], f[idx], 'ro')
plt.show()
idx
```
